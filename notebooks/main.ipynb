```python
columns = []

for i in list(df9):

    if i == 'Price (in USD)':
        continue 
    else:
      columns.append(i)
```

```python
def get_car_prediction(make, year, engine_size, hp, torque, time_0_60):

    mapping = {
        'Year': year,
        'Engine Size (L)': engine_size,
        'Horsepower': hp,
        'Torque (lb-ft)': torque,
        '0-60 MPH Time (seconds)': time_0_60
    }

    user_make = f"Car Make_{make.strip().lower()}"

    input_row = {col: 0 for col in columns}

    for key, value in mapping.items():
        input_row[key] = value

    if user_make in input_row:
        input_row[user_make] = 1

    df_input = pd.DataFrame([input_row])[columns]

    prediction = model.predict(df_input)[0]

    return prediction


```
